```python
train_X = [string.lower() for string in df["text"]]
```

```python
import re
import sys
```

```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
train_X = [remove_emoji(string) for string in train_X]
train_X
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
ls=[]
ls.append([morph.parse(line)[0].normal_form for line in train_X])
print(ls)
```

```python
import collections
total=[]
line=ls[0]
for word in line: 
    c = collections.Counter()
    for a in word.split(' '):
        c[a] += 1
        
    total.append(list(c)) 
```

```python
total
```

```python
id_list=[]
for l in total:
    #print(l)
    if "nan" in l:
        id_list.append('0')
    elif ("—Å–∫–∏–¥–∫–∞" in l) or ("—Å–∫–∏–¥–∫–∏" in l):
        id_list.append('1')
    elif ("–¥–æ—Å—Ç–∞–≤–∏–º" in l) or ("–¥–æ—Å—Ç–∞–≤–∫–∞" in l):   
        id_list.append('2')
    elif "–∫–æ—Ä–∑–∏–Ω–µ" in l:   
        id_list.append('3')
    else: 
        id_list.append('4')
    
#print(len(id_list))  
```

```python
print(id_list) 
```

```python
df=df.assign(text=id_list)
```

```python
df=df.drop(columns=['subject', 'body'])
```

```python
df.to_csv('new_actions.csv',index = False, header=True)
```
